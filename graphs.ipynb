## The depressive symptoms presence distribution
### Before removing subjectivity

```python
import pandas as pd
%matplotlib inline

import matplotlib.pyplot as plt
plt.figure()
plt.rcParams.update({'font.size': 22})

SYMPTOM_COLUMN_NAME_LIST = ["Diminished interest",
                            "Depressed mood",
                            "Sleep troubles",
                            "Fatigue",
                            "Appetite problems",
                            "Worthlessness",
                            "Inability to concentrate",
                            "Psychomotor activity",
                            "Suicidal thoughts"
                            ]

df = pd.read_csv('stats/symptom_bin_dist.csv')
df.rename(columns={'no_count': 'No', 'yes_count': 'Yes'}, inplace=True)
ax = df.plot(x='symptom', y=['No', 'Yes'], kind='bar', color=['green', 'red'], figsize=(20, 10), fontsize=20)
ax.set_xticklabels(SYMPTOM_COLUMN_NAME_LIST, rotation = 60)
ax.set_title("Original distribution", pad=20)
ax.set_xlabel("Depressive symptom", labelpad=20)
ax.set_ylabel("Count", labelpad=20)
for container in ax.containers:
    ax.bar_label(container, fontsize=14, padding=6)
```

## The depressive symptoms presence distribution
### After removing subjectivity

```python
df = pd.read_csv('stats/symptom_bin_nosub_dist.csv')
df.rename(columns={'no_count': 'No', 'yes_count': 'Yes'}, inplace=True)
ax = df.plot(x='symptom', y=['No', 'Yes'], kind='bar', color=['green', 'red'], figsize=(20, 10), fontsize=20)
ax.set_xticklabels(SYMPTOM_COLUMN_NAME_LIST, rotation = 60)
ax.set_title("Distribution after removing subjectivity", pad=20)
ax.set_xlabel("Depressive symptom")
ax.set_ylabel("Count")
ax.set_xlabel("Depressive symptom", labelpad=20)
ax.set_ylabel("Count", labelpad=20)
for container in ax.containers:
    ax.bar_label(container, fontsize=14, padding=6)
```
